## 8. 딥러닝
### 8.1 더 깊게
#### 8.1.1 더 깊은 신경망으로

#### - 3 x 3의 작은 필터를 사용한 합성곱 계층
#### - ReLU / 완전연결 계층 뒤에 드롭아웃 / Adam / He 초기값

#### 8.1.2 정확도를 더 높이려면
#### - Data Augmentation(데이터 확장) : 입력 이미지를 알고리즘을 동원해 인위적으로 확장, 입력 이미지를 회전하거나 세로로 이동하는 등 미세한 변화를 주어 이미지 개수를 증가, 이미지를 잘라내는 crop, 좌우 뒤집는 flip, 밝기 등의 외형 변화, 확대, 축소

#### 8.1.3 깊게 하는 이유
#### - '층을 깊게 하는 것'이 왜 중요한가에 대한 이론적인 근거는 아직 부족
#### - 층을 깊게 햇을 때 신경망의 매개변수 수가 줄어든다 = 층을 깊게 한 신경망은 깊지 않은 경우보다 적은 매개변수로 같은 수준의 표현력을 달성 
#### - 5x5 합성곱 연산 1회 = 3x3 합성곱 연산 2회

#### - 학습의 효율성 증대

### 8.2 딥러닝의 초기 역사
#### 8.2.1 이미지넷
#### 8.2.2 VGG
#### - 합성곱 계층 + 풀링 계층

#### 8.2.3 GoogLeNet
#### - 인셉션 구조 : 가로 방향에 '폭' 존재 -> 크기가 다른 필터, 풀링을 여러 개 적용하여 결과를 결합


#### 8.2.4 ResNet
#### Skip Connection(스킵 연결) : 층을 깊게 할 수 있는 특별한 장치, 입력 데이터를 합성곱 계층을 건너뛰어 출력에 바로 더하는 구조, 역전파 때 신호 감쇠를 방지



### 8.3 딥러닝 고속화 
#### - GPU : 대량 병렬 연산에 장점 / CPU : 연속적인 복잡한 계산에 장점
#### - CUDA : 엔비디아의 GPU 컴퓨팅용 통합 개발 환경 / cuDNN : CUDA 위에서 동작하는 라이브러리로 딥러닝에 최적화된 함수 등이 구현
#### - 딥러닝은 높은 수치 정밀도(수치를 몇 비트로 표현하느냐)를 요구 하지 않음 - 신경망의 중요한 성질 중 하나로, 신경망의 견고성에 따른 특성
#### - 딥러닝을 고속화하기 위해 비트를 줄이는 기술은 점차 중요해짐

### 8.4 딥러닝의 활용
#### 8.4.1 사물 검출(Object Detection)
#### - R-CNN(Regions with CNN)
#### 8.4.2 분할(Segmentation)
#### - FCN(Fully Convlutional Network) : 합성곱 계층만으로 구성된 네트워크

#### 8.4.3 사진 캡션 생성
#### - NIC(Neural Image Cation) : RNN으로 구성
#### - Multimodal processing : 여러 종류의 정보를 조합하고 처리
